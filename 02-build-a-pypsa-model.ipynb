````python
import pypsa
import pandas as pd
pypsa.options.api.new_components_api = True

n = pypsa.Network()
````

````python
n.set_snapshots(pd.date_range('2025-01-01', '2025-12-31 23:00', freq='h'))
````

````python
n.add('Carrier',['gas','AC'])
````

````python
n.add('Bus','Location',carrier = 'AC')
````

````python
n.add('Generator',
      'Generator A',
      bus = 'Location', 
      p_nom = 500, 
      marginal_cost = 1,
      p_nom_extendable = True
      )
````

````python
n.add('Generator',
      'Generator A',
      bus = 'Location', 
      p_nom = 500, 
      marginal_cost = 1,
      p_nom_extendable = False
      )
````

````python
n.add('Load',
      'Load A',
      bus = 'Location',
      p_set = 100,
      carrier = 'gas')
````

````python
n.add('Load',
      'Load B',
      bus = 'Location',
      p_set = 100,
      carrier = 'gas')
````
