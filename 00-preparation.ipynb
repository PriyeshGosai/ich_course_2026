Collections

````python
# List variable
list_var = [1, 2, 3, 4, 5]
print(list_var)

# Accessing specific items
print(list_var[0])        # First item
print(list_var[-1])       # Last item
print(list_var[1:3])      # Slice (items 1 to 2)

# Performing operations
list_var.append(6)        # Add item to end
list_var.insert(0, 0)     # Insert at position
list_var.remove(3)        # Remove specific item
list_var.pop()            # Remove last item
print(len(list_var))      # Length of list
print(sum(list_var))      # Sum of items
print(max(list_var))      # Maximum value
print(min(list_var))      # Minimum value
````

````python
# Dictionary variable
dict_var = {'name': 'Alice', 'age': 30, 'city': 'Johannesburg'}
print(dict_var)

# Accessing specific items
print(dict_var['name'])           # Access by key
print(dict_var.get('age'))        # Access with get method
print(dict_var.keys())            # Get all keys
print(dict_var.values())          # Get all values
print(dict_var.items())           # Get all key-value pairs

# Performing operations
dict_var['age'] = 31              # Update value
dict_var['country'] = 'South Africa'  # Add new key-value pair
dict_var.pop('city')              # Remove key-value pair
print(len(dict_var))              # Number of key-value pairs
print('name' in dict_var)         # Check if key exists
````

````python
# Set variable
set_var = {1, 2, 3, 4, 5}
print(set_var)

# Accessing specific items (sets are unordered, no indexing)
print(3 in set_var)               # Check if item exists

# Performing operations
set_var.add(6)                    # Add item
set_var.remove(3)                 # Remove item (error if not found)
set_var.discard(2)                # Remove item (no error if not found)
print(len(set_var))               # Number of items

# Set operations
set_var2 = {4, 5, 6, 7}
print(set_var.union(set_var2))    # Combine sets
print(set_var.intersection(set_var2))  # Common items
print(set_var.difference(set_var2))    # Items in set_var but not set_var2
````

````python
# Tuple variable
tuple_var = (1, 2, 3, 4, 5)
print(tuple_var)

# Accessing specific items
print(tuple_var[0])               # First item
print(tuple_var[-1])              # Last item
print(tuple_var[1:3])             # Slice (items 1 to 2)

# Performing operations (tuples are immutable, limited operations)
print(len(tuple_var))             # Length of tuple
print(tuple_var.count(3))         # Count occurrences of item
print(tuple_var.index(3))         # Index of first occurrence
print(sum(tuple_var))             # Sum of items
print(max(tuple_var))             # Maximum value
print(min(tuple_var))             # Minimum value

# Note: Cannot modify tuples (no append, remove, etc.)
````

**Heat Rate and Efficiency**

Heat rate is a measure of how much thermal energy (in BTU) is required to generate one kilowatt-hour (kWh) of electricity. It is the inverse of efficiency and is commonly used by engineers in the power generation industry.

**Relationship to Efficiency**

Efficiency and heat rate are inversely related:

Efficiency (%) = 3,412.14 / Heat Rate (BTU/kWh) × 100

The constant 3,412.14 represents the number of BTU in one kWh.

**Why Use Heat Rate?**

- Practical application: Engineers prefer heat rate because it directly reflects fuel consumption—higher heat rate means more fuel is needed to produce the same electricity
- Operational focus: Power plant operators monitor heat rate to assess plant performance and operating costs
- Lower is better: A lower heat rate indicates better efficiency and lower fuel costs
- Industry standard: Heat rate is the standard metric in power generation for comparing generator performance


Energy Cost Calculation Equations

1. Heat Rate Conversion (Imperial to Metric)

$$HR_{met} = HR_{imp} \times 0.2931$$

where:
- $HR_{met}$ = Heat rate in MJ/kWh
- $HR_{imp}$ = Heat rate in Btu/kWh
- 0.2931 = Conversion factor

2. Efficiency

$$\eta = \frac{1}{HR_{met}}$$

where:
- $\eta$ = Thermal efficiency (kWh/MJ)
- $HR_{met}$ = Heat rate in MJ/kWh

3. Heat Input

$$Q_{in} = \frac{P_{nom}}{\eta}$$

where:
- $Q_{in}$ = Heat input required (MJ/h)
- $P_{nom}$ = Nominal power output (MW)
- $\eta$ = Thermal efficiency

4. Coal Flow Rate

$$\dot{m}_{coal} = \frac{Q_{in}}{CV} \times \frac{3600}{1000}$$

where:
- $\dot{m}_{coal}$ = Coal mass flow rate (tonnes/h)
- $Q_{in}$ = Heat input required (MJ/h)
- $CV$ = Calorific value of coal (MJ/kg)
- 3600/1000 = Conversion factor (s/h ÷ kg/tonne)

5. Marginal Cost

$$C_{marginal} = \dot{m}_{coal} \times c_{coal}$$

where:
- $C_{marginal}$ = Marginal operating cost (ZAR/MWh)
- $\dot{m}_{coal}$ = Coal mass flow rate (tonnes/h)
- $c_{coal}$ = Coal unit cost (ZAR/tonne)

Number Formatting in Python

`:.4f` is a format specifier that controls how numbers are displayed.

- `{variable}` - the variable to display
- `:` - starts the format specification
- `.4` - display 4 decimal places
- `f` - format as a floating-point number

````python
def linear_func(x, a, b):
    answer = a * x + b
    return answer


heat_rate_50_MW = [13.396, 12.173, 11.441, 11.103, 11.019] 

load = [0.25, 0.4, 0.6, 0.8, 1]

fitted_curve, _ = curve_fit(linear_func, load, heat_rate_50_MW)
````

````python
a, b = fitted_curve

print(f"Fitted Parameters:")
print(f"a (slope): {a:.4f}")
print(f"b (intercept): {b:.4f}")
print()
print(f"Fitting Equation:")
print(f"Heat Rate = {a:.4f} × Load + {b:.4f}")
print()
print(f"Example Predictions:")
for load_value in load:
    predicted_heat_rate = linear_func(load_value, a, b)
    print(f"Load: {load_value} → Heat Rate: {predicted_heat_rate:.4f} BTU/kWh")
````

**Note:** 

The `curve_fit` function returns two outputs: the polynomial coefficients and the covariance matrix. <br> 
For this course, we only need the coefficients. Therefore, we use an underscore `_` to ignore the covariance matrix output.

Example:
```python
coefficients, _ = curve_fit(function, x_data, y_data)
# coefficients contains the fitted parameters
# _ discards the covariance matrix (we don't need it)
```
